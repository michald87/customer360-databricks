```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.erp;
```

```python
%sql
DROP TABLE IF EXISTS workspace.erp.customer;

CREATE OR REPLACE TABLE workspace.erp.customer (
  customer_id BIGINT COMMENT 'Unique customer identifier',
  first_name STRING COMMENT 'Customer first name',
  last_name STRING COMMENT 'Customer last name',
  pesel STRING COMMENT 'Polish national identification number',
  birth_date DATE COMMENT 'Date of birth',
  email STRING COMMENT 'Email address',
  phone STRING COMMENT 'Phone number',
  address STRING COMMENT 'Street address',
  city STRING COMMENT 'City',
  postal_code STRING COMMENT 'Postal/ZIP code',
  country STRING COMMENT 'Country',
  registration_date TIMESTAMP COMMENT 'Account registration date',
  customer_status STRING COMMENT 'Customer status (active, inactive, suspended)',
  last_purchase_date DATE COMMENT 'Date of last purchase',
  total_purchases DECIMAL(10,2) COMMENT 'Total purchase amount'
);
```
